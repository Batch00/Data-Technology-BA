# Introduction to BeautifulSoup

```python
# Import BeautifulSoup
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'http://quotes.toscrape.com/'
page = requests.get(url)
soup = bs(page.content, 'html.parser')
```

```python
print(soup)
```

```python
# Scrape the heading
h2 = soup.find('h2')
heading = h2.text
print(h2)
print(heading)
```

```python
# Scrape the top ten tags
tag_box = soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
url = 'https://realpython.github.io/fake-jobs/'
page = requests.get(url)
soup = bs(page.content, 'html.parser')
```

```python
job_box = soup.find('div', class_='columns is-multiline')


job_title = job_box.find_all('h2', class_='title is-5')
for job in job_title:
    word = job.text
    print(word)
```

```python
employers = job_box.find_all('h3', class_='subtitle is-6 company')
for employer in employers:
    word = employer.text
    print(word)
```

```python
locations = job_box.find_all('p', class_='location')
for loc in locations:
    word = loc.text
    print(word)
```

```python
apply_list = []
apply_buttons = job_box.find_all('a', class_='card-footer-item')
for a in apply_buttons:
    link = a['href']
    if a.text == "Apply":
        apply_list.append(link)
        print(link)
```

```python
job_description_list = []

for i in range(len(apply_list)):
    job_page = requests.get(apply_list[i])
    job_soup = bs(job_page.content, 'html.parser')
    content = job_soup.find('div', class_="content")
    job_description_list.append(content.text)
    
print(job_description_list[0])
```
